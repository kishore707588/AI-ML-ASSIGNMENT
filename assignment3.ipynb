```python
# TASK-5 Build ML model with linear regression (Target column is price)
```

Linear Regression model is in form Y = MX+C

```python
df = df[['price','sqft_lot']]
df
```

Independent Variable

```python
X = df['sqft_lot']
```

Dependent Variale

```python
Y = df['price']
```

Main Process

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,Y)
plt.xlabel('sqft-lot')
plt.ylabel('price')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.4,random_state = 23)
```

```python
X_train = np.array(X_train).reshape(-1,1)
```

```python
X_train
```

```python
X_test = np.array(X_test).reshape(-1,1)
X_test
```

Fitting of Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
c = lr.intercept_
c
```

```python
M = lr.coef_
M
```

Since we have M,C values lets try predicting in Y=MX+C

```python
Y_pred_train = M*x_train+c
Y_pred_train.flatten()
```

```python
Y_pred_train1 = lr.predict(x_train)
Y_pred_train1
```

```python
plt.scatter(X_train,Y_train)
plt.scatter(X_train,Y_pred_train1,color = 'red')
plt.xlabel('sqft-lot')
plt.ylabel('price')
```

```python
plt.scatter(X_train,Y_train)
plt.plot(X_train,Y_pred_train1,color = 'red')
plt.xlabel('sqft-lot')
plt.ylabel('price')
```

Now let's try fitting the model for X_test,Y_test

```python
lr.fit(X_test,Y_test)
```
